Notebook for showing that the 2-phase X-SH code does not work for uneven numbers of donors and acceptors. I do this by comparing the diabatic populations for a 3x3 and 3x4 6T-PDI system, which shows that the same diabatic states (in terms of index) are exchanging population in each system. This means that the wrong CT-diabat is coupled to the exciton in the 3x4 case, as its index should refer to a non-interfacial donor molecule. In both cases, the CT-couplings and XT-couplings were turned off, and the test was conducted with Wei-Tao's version of X-SH (before I made any changes) and the version I used for the 6T-PDI 1D paper.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tag_3x3 = '3x3'
tag_3x4 = '3x4'
nuclear_timestep = 0.1

populations_3x3 = np.loadtxt(f'{tag_3x3}/{tag_3x3}_diabat_populations.txt')
populations_3x4 = np.loadtxt(f'{tag_3x4}/{tag_3x4}_diabat_populations.txt')

populations_3x3_fi = np.loadtxt(f'{tag_3x3}/{tag_3x3}_fi_diabat_populations.txt')
populations_3x4_fi = np.loadtxt(f'{tag_3x4}/{tag_3x4}_fi_diabat_populations.txt')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, rasterized=True, dpi=400, figsize=(8,3.5))

ax[0].plot(np.arange(len(populations_3x3[:,0]))*nuclear_timestep, populations_3x3[:,9], color='purple', linewidth=2, label='State 10 - v4.1')
ax[0].plot(np.arange(len(populations_3x3[:,0]))*nuclear_timestep, populations_3x3[:,3], color='c', linewidth=2, label='State 4 - v4.1')

ax[0].plot(np.arange(len(populations_3x3_fi[:,0]))*nuclear_timestep, populations_3x3_fi[:,9], color='orange', linewidth=1, linestyle='--', label='State 10 - filip_barrier')
ax[0].plot(np.arange(len(populations_3x3_fi[:,0]))*nuclear_timestep, populations_3x3_fi[:,3], color='r', linewidth=1, linestyle='--', label='State 4 - filip_barrier')
ax[0].set_title('Populations \n Diabats 4 and 10 \n 3x3 System')
ax[0].legend(frameon=False)

ax[1].plot(np.arange(len(populations_3x4[:,0]))*nuclear_timestep, populations_3x4[:,12], color='purple', linewidth=2, label='State 13 - v4.1')
ax[1].plot(np.arange(len(populations_3x4[:,0]))*nuclear_timestep, populations_3x4[:,3], color='c', linewidth=2, label='State 4 - v4.1')
ax[1].plot(np.arange(len(populations_3x4[:,0]))*nuclear_timestep, populations_3x4[:,4], color='pink', linewidth=2, label='State 5 - v4.1')


ax[1].plot(np.arange(len(populations_3x4_fi[:,0]))*nuclear_timestep, populations_3x4_fi[:,12], color='orange', linewidth=1, linestyle='--', label='State 13 - filip_barrier')
ax[1].plot(np.arange(len(populations_3x4_fi[:,0]))*nuclear_timestep, populations_3x4_fi[:,3], color='r', linewidth=1, linestyle='--', label='State 4 - filip_barrier')
ax[1].plot(np.arange(len(populations_3x4_fi[:,0]))*nuclear_timestep, populations_3x4_fi[:,4], color='lime', linewidth=1, linestyle='--', label='State 5 - filip_barrier')
ax[1].set_title('Populations \n Diabats 4,5 and 13 \n 3x4 System')
ax[1].legend(frameon=False)

fig.supylabel('Population')
fig.supxlabel('Time (fs)')
fig.tight_layout()
```
