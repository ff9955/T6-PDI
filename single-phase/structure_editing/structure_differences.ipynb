Notebook where I plot the distributions of bond lengths and internal angles of atoms in L8B0 derivatives, each unique bond/angle is plotted as an average over the 8 molecules in the L8B0 in the unit cell and compared to the associated standard deviation.

Checks: printed lists of atom types to check that correct atom types are being assigned to each list, also checked arN/arS calculated angles of molecules 1 and 2 in unit cell, they match up well with vmd angles.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_labels = ['arC', 'arN', 'arS', 'al']
mol_tag = 'L8B0'
```

```python
aromatic_C_angles = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[0]}-internal-angles.txt')
aromatic_N_angles = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[1]}-internal-angles.txt')
aromatic_S_angles = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[2]}-internal-angles.txt')
aliphatic_angles = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[3]}-internal-angles.txt')

aromatic_C_bonds = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[0]}-bonds.txt')
aromatic_N_bonds = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[1]}-bonds.txt')
aromatic_S_bonds = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[2]}-bonds.txt')
aliphatic_bonds = np.loadtxt(f'L8B0_bonds_angles/{mol_tag}-{array_labels[3]}-bonds.txt')
```

```python
arC_bond_mean, arC_bond_std = np.mean(aromatic_C_bonds, axis=0), np.std(aromatic_C_bonds, axis=0)
arN_bond_mean, arN_bond_std = np.mean(aromatic_N_bonds, axis=0), np.std(aromatic_N_bonds, axis=0)
arS_bond_mean, arS_bond_std = np.mean(aromatic_S_bonds, axis=0), np.std(aromatic_S_bonds, axis=0)
al_bond_mean, al_bond_std = np.mean(aliphatic_bonds, axis=0), np.std(aliphatic_bonds, axis=0)

arC_angle_mean, arC_angle_std = np.mean(aromatic_C_angles, axis=0), np.std(aromatic_C_angles, axis=0)
arN_angle_mean, arN_angle_std = np.mean(aromatic_N_angles, axis=0), np.std(aromatic_N_angles, axis=0)
arS_angle_mean, arS_angle_std = np.mean(aromatic_S_angles, axis=0), np.std(aromatic_S_angles, axis=0)
al_angle_mean, al_angle_std = np.mean(aliphatic_angles, axis=0), np.std(aliphatic_angles, axis=0)
```

```python
plt.scatter(arC_bond_mean, arC_bond_std, s=10, color='k', label='aromatic C')
plt.scatter(arN_bond_mean, arN_bond_std, s=10, color='cyan', label='aromatic N')
plt.scatter(arS_bond_mean, arS_bond_std, s=10, color='orange', label='aromatic S')
plt.scatter(al_bond_mean, al_bond_std, s=10, color='gray', label='aliphatic')
plt.ylabel('Standard deivation (\u212B)')
plt.xlabel('Mean bond length (\u212B)')
plt.legend()
plt.title('Mean bond length vs Standard deviation')
```

```python
plt.scatter(arC_angle_mean, arC_angle_std, s=10, color='k', label='Aromatic C')
plt.scatter(arN_angle_mean, arN_angle_std, s=10, color='cyan', label='Aromatic N')
plt.scatter(arS_angle_mean, arS_angle_std, s=10, color='orange', label='Aromatic S')
plt.scatter(al_angle_mean, al_angle_std, s=10, color='gray', label='Aliphatic')
plt.ylabel('Standard deviation')
plt.xlabel('Mean bond angle (Deg.)')
plt.title('Mean bond angle vs Standard deviation')
plt.xlim([75,190])
plt.legend()
```
