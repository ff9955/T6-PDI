This notebook is for checking distance between the centres of mass of T6 and PCTDI molecules in Sam's problematic 3x3 simulation, where the overlaps diverge after 7 time steps. The distances should be less than 12 Angstroms for molecules that have non-zero coupling in the run.log file of the simulation, since the AOM section of the input file indicates a distance cutoff of 12 Angstroms.

```python
import sys
sys.path.append('/homes/fivanovic/Work/FOB-SH/')
```

```python
import FOB_SH_analysis_Filip as FB
import numpy as np
```

```python
three_five_coords = FB.data_splitter('/homes/fivanovic/Work/X-SH/AOM_debugging/Filip_checks/problem_dimers/real_run_step7/real_problem_285_318.xyz', 4)

three_five_coords = np.array(three_five_coords)
vector_float = np.vectorize(float)

for row in three_five_coords:
    
    if row[0] == 'C':
        row[1:] = 12*vector_float(row[1:])
    elif row[0] == 'S':
        row[1:] = 32*vector_float(row[1:])
    elif row[0] == 'O':
        row[1:] = 16*vector_float(row[1:])
    elif row[0] == 'N':
        row[1:] = 14*vector_float(row[1:])

print(three_five_coords)

three_five_coords = three_five_coords[:, 1:]
three_five_coords = vector_float(three_five_coords)

three_coords = three_five_coords[0:44]
five_coords = three_five_coords[44:]
```

```python
T6_Mr = 494
PCTDI_Mr = 390
```

```python
three_CoM = np.sum(three_coords, axis=0)/T6_Mr
five_CoM = np.sum(five_coords, axis=0)/PCTDI_Mr

print(three_CoM)
print(five_CoM)

three_five_distance = np.sqrt(np.sum((three_CoM - five_CoM)**2))
three_five_distance
```

```python
two_six_coords = FB.data_splitter('/homes/fivanovic/Work/X-SH/AOM_debugging/Filip_checks/problem_dimers/real_run_step7/real_problem_284_330.xyz', 4)

two_six_coords = np.array(two_six_coords)

for row in two_six_coords:
    
    if row[0] == 'C':
        row[1:] = 12*vector_float(row[1:])
    elif row[0] == 'S':
        row[1:] = 32*vector_float(row[1:])
    elif row[0] == 'O':
        row[1:] = 16*vector_float(row[1:])
    elif row[0] == 'N':
        row[1:] = 14*vector_float(row[1:])

two_six_coords = two_six_coords[:, 1:]
two_six_coords = vector_float(two_six_coords)

two_coords = two_six_coords[0:44]
six_coords = two_six_coords[44:]
```

```python
two_CoM = np.sum(two_coords, axis=0)/T6_Mr
six_CoM = np.sum(six_coords, axis=0)/PCTDI_Mr

two_six_distance = np.sqrt(np.sum((two_CoM - six_CoM)**2))
two_six_distance
```
