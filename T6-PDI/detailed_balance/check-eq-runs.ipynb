Here, I'm just plotting the temperature fluctuations for the NVT and NVE equilibration runs for T6:PDI. This is to check that the system equilibrates properly.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('/scratch/fivanovic/X-SH/T6-PDI/detailed_balance/equilibration_files/NVT/run-1.ener') as ener_file:

    all_lines = ener_file.readlines()
    all_lines = all_lines[1:]

    all_lines = [l.split() for l in all_lines]
    all_lines = [list(map(float, l1)) for l1 in all_lines]


kinetic_energies = [l[2] for l in all_lines]
potential_energies = [l[4] for l in all_lines]
temperature = [l[3] for l in all_lines]
simulation_time = [l[0]*0.1 for l in all_lines]
```

```python
plt.plot(simulation_time, kinetic_energies, color = 'orange', label = 'Kinetic Energy')
plt.plot(simulation_time, potential_energies, color = 'c', label = 'Potential Energy')
plt.xlabel('Timestep /fs')
plt.ylabel('Energy /Hartrees')
plt.title('NVT Run - Energy vs Time')
plt.legend()
```

```python
plt.plot(simulation_time[:25000], kinetic_energies[:25000], color = 'orange', label = 'Kinetic Energy')
plt.plot(simulation_time[:25000], potential_energies[:25000], color = 'c', label = 'Potential Energy')
plt.xlabel('Timestep /fs')
plt.ylabel('Energy /Hartrees')
plt.title('NVT Run - Energy vs Time')
plt.legend()
```

```python
plt.plot(simulation_time, temperature, color = 'red')
plt.xlabel('Timestep /fs')
plt.ylabel('Temperature /K')
plt.title('NVT Run - Temperature vs Time')
```

```python
plt.plot(simulation_time[:250000], temperature[:250000], color = 'red')
plt.xlabel('Timestep /fs')
plt.ylabel('Temperature /K')
plt.title('NVT Run - Temperature vs Time')
```

```python
with open('/scratch/fivanovic/X-SH/T6-PDI/detailed_balance/equilibration_files/NVE/run-1.ener') as ener_file:

    all_lines = ener_file.readlines()
    all_lines = all_lines[1:]

    all_lines = [l.split() for l in all_lines]
    all_lines = [list(map(float, l1)) for l1 in all_lines]


kinetic_energies = [l[2] for l in all_lines]
potential_energies = [l[4] for l in all_lines]
temperature = [l[3] for l in all_lines]
simulation_time = [l[0]*0.1 for l in all_lines]

print('Average NVE run temperature for first 20ps is:', sum(temperature[:200000])/len(temperature[:200000]))
print('Average NVE run temperature for last 10ps is:', sum(temperature[300000:])/len(temperature[300000:]))

acc_temp_avg = np.cumsum(temperature) / np.arange(1, len(temperature) + 1)
```

```python
plt.plot(simulation_time, temperature, color = 'red', label = 'Temperature')
plt.plot(simulation_time, acc_temp_avg, color = 'c', linestyle = '--', label = 'Accumulated average')
plt.xlabel('Timestep /fs')
plt.ylabel('Temperature /K')
plt.title('NVE Run - Temperature vs Time')
plt.ylim([270,330])
plt.legend()
```

```python
plt.plot(simulation_time[:25000], temperature[:25000], color = 'red', label = 'Temperature')
plt.plot(simulation_time[:25000], acc_temp_avg[:25000], color = 'c', linestyle = '--', label = 'Accumulated average')
plt.xlabel('Timestep /fs')
plt.ylabel('Temperature /K')
plt.title('NVE Run - Temperature vs Time')
plt.ylim([270,330])
plt.legend()
```
