```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def linear_function(x, gradient):
    return gradient*x
```

```python
timestep_length = np.array([0.32, 0.44, 0.86, 1.45, 1.9, 2.8, 3.1, 7.5, 11.7, 18.75])
number_states = np.array([12, 30, 100, 225, 400, 576, 625, 900, 1225, 1600])

optimised_gradient, extra = curve_fit(linear_function, number_states, timestep_length)
predicted_lengths = number_states*optimised_gradient

plt.plot(number_states, timestep_length, color='b', linewidth=2, linestyle='--')
plt.scatter(number_states, timestep_length, color='r', s=20, marker='x')
#plt.plot(number_states, predicted_lengths, color='plum')
plt.ylabel('Timestep length (s)')
plt.xlabel('Number of States')
plt.title('Length of timestep vs number of states')
```
