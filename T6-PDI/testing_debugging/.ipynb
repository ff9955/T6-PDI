```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
XSH_PDI_coupling_dir = '/scratch/fivanovic/X-SH/testing_debugging/POD_T6_PDI_TEST/run-fssh-short/X-SH_dimer_coupling/PDI_dimer_coupling.xyz'
with open(XSH_PDI_coupling_dir) as full_file:
    XSH_PDI_coupling_list = full_file.readlines()

    XSH_PDI_coupling_list = map(float, XSH_PDI_coupling_list)
    XSH_PDI_coupling_list = list(XSH_PDI_coupling_list)
    print(XSH_PDI_coupling_list)

XSH_PDI_coupling_array = np.array(XSH_PDI_coupling_list)*27200

XSH_PDI_mean = np.mean(XSH_PDI_coupling_array)
XSH_PDI_sd = np.std(XSH_PDI_coupling_array)
```

```python
XSH_T6_coupling_dir = '/scratch/fivanovic/X-SH/testing_debugging/POD_T6_PDI_TEST/run-fssh-short/X-SH_dimer_coupling/T6_dimer_coupling.xyz'
with open(XSH_T6_coupling_dir) as full_file:
    XSH_T6_coupling_list = full_file.readlines()

    XSH_T6_coupling_list = map(float, XSH_T6_coupling_list)
    XSH_T6_coupling_list = list(XSH_T6_coupling_list)
    print(XSH_T6_coupling_list)

XSH_T6_coupling_array = np.array(XSH_T6_coupling_list)*27200

XSH_T6_mean = np.mean(XSH_T6_coupling_array)
XSH_T6_sd = np.std(XSH_T6_coupling_array)
```

```python
plt.hist(XSH_PDI_coupling_array, color = 'k', alpha = 0.5)
plt.xlabel('Electronic Coupling /meV')
plt.ylabel('Frequency')
plt.title('X-SH PDI-PDI Coupling Distribution')
plt.text(25, 15.5, f'Mean: {XSH_PDI_mean:2f}')
plt.text(25, 14, f'\u03C3: {XSH_PDI_sd:2f}')
```

```python
plt.hist(XSH_T6_coupling_array, color = 'r', alpha = 0.3)
plt.xlabel('Electronic Coupling /meV')
plt.ylabel('Frequency')
plt.title('X-SH T6-T6 Coupling Distribution')
plt.text(40, 16.5, f'Mean: {XSH_T6_mean:2f}')
plt.text(40, 15, f'\u03C3: {XSH_T6_sd:2f}')
```

```python
timesteps = np.arange(1,102)*10
```

```python
plt.plot(timesteps, XSH_PDI_coupling_array, color = 'k')
plt.xlabel('Time /fs')
plt.ylabel('PDI-PDI Coupling /meV')
```

```python
plt.plot(timesteps, XSH_T6_coupling_array, color = 'r')
plt.xlabel('Time /fs')
plt.ylabel('T6-T6 Coupling /meV')
```
